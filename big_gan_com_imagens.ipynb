<h3>BIG GAN na Geração de Imagens em Alta Resolução (Simulando a Terraformação de Marte)</h3>

<h4>Importação das Classes e Declaração das Variáveis</h4>

```python
from Neuraline.ArtificialIntelligence.DeepLearning.generative_adversarial_networks import GenerativeAdversarialNetworks
generative_adversarial_networks = GenerativeAdversarialNetworks()
url_path, url_path1, url_path2 = './planetas', './planetas/earth.jpg', './planetas/mars.jpg'
```

<h4>Primeira Forma: Sem Treinamento</h4>

```python
print('[gerando imagem, aguarde...]')
result = generative_adversarial_networks.predictBIGGAN(url_path1=url_path1, url_path2=url_path2, progress=False)
if result: print('[imagem gerada com sucesso.]')
else: print('[erro na geração da imagem.]')
```

<h4>Segunda Forma: Com Treinamento</h4>

```python
result1 = generative_adversarial_networks.addFit(url_path)
result2 = generative_adversarial_networks.saveModel('terraformacao_com_big_gan')
if result1 and result2: print('[...modelo treinado com sucesso]')
else: print('[erro no treinamento do modelo.]')
```

```python
print('[gerando imagem, aguarde...]')
result1 = generative_adversarial_networks.loadModel('terraformacao_com_big_gan')
result2 = generative_adversarial_networks.predictBIGGAN(url_path1=0, url_path2=1, progress=False)
if result1 and result2: print('[imagem gerada com sucesso.]')
else: print('[erro na geração da imagem.]')
```
